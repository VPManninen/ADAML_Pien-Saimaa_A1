Comparing the prediction horizon of three different models. 
MAE losses from every test sample have been stored in a file so they can be plotted as bar plot for easy compare

```python
import numpy as np
import matplotlib.pyplot as plt

# empty list to read list from a file
tnn_losses = []

# open file and read the content in a list
with open("TNN_losses.txt", 'r') as fp:
    index = 1
    x = 0
    for line in fp:
        # remove linebreak from a current name
        # linebreak is the last character of each line
        # compute the mean absolute error from the absolute errors stored in the testing phase
        new_val = float(line[:-1])**2
        x = x  + new_val/index
        index = index + 1
        # add current item to the list
        tnn_losses.append(x)



# empty list to read list from a file
rnn_losses = []

# open file and read the content in a list
with open("RNN_losses.txt", 'r') as fp:
    index = 1
    x = 0
    for line in fp:
        # remove linebreak from a current name
        # linebreak is the last character of each line
        new_val = float(line[:-1])**2
        x = x  + new_val/index
        index = index + 1

        # add current item to the list
        rnn_losses.append(x)


# empty list to read list from a file
lstm_losses = []

# open file and read the content in a list
with open("LSTM_losses.txt", 'r') as fp:
    index = 1
    x = 0
    for line in fp:
        # remove linebreak from a current name
        # linebreak is the last character of each line
        new_val = float(line[:-1])**2
        x = x  + new_val/index
        index = index + 1

        # add current item to the list
        lstm_losses.append(x)
```

```python
# set width of bar 
barWidth = 0.25
fig, ax = plt.subplots(figsize =(12, 8)) 


# Set position of bar on X axis 
br1 = np.arange(len(tnn_losses[:50])) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 

# Make the plot
plt.bar(br1, tnn_losses[0:50], color ='r', width = barWidth, 
		edgecolor ='grey', label ='TNN') 
plt.bar(br2, lstm_losses[0:50], color ='b', width = barWidth, 
		edgecolor ='grey', label ='LSTM') 
plt.bar(br3, rnn_losses[0:50], color ='g', width = barWidth, 
		edgecolor ='grey', label ='RNN') 


plt.xlabel('x steps ahead prediction', fontweight ='bold', fontsize = 15) 
plt.ylabel('MSE value', fontweight ='bold', fontsize = 15)
labels = np.arange(0, 0.0035, step=0.0005).astype(str)
ax.set_ylim(bottom=0, top=0.0035)
ax.set_yticks(np.arange(0, 0.0035, step=0.0005), labels=labels)

plt.grid(True)
plt.xticks(fontsize=8)
plt.yticks(fontsize=14)
plt.legend()
plt.show() 
```
